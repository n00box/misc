```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.random((10,3)),
                  columns = ['A', 'B', 'C'])
print(df)
def _update_min(arr):
    mini = arr.min()  # get minimum value
    df['A'][arr.index[-1]] = mini  # update our dataframe
    return mini
df["A"].rolling(2, min_periods = 1).apply(_update_min)
df
```

```python
df = pd.DataFrame(np.random.random((10,3)),
                  columns = ['A', 'AA', 'C'])
df.AA = df.A
df["A"].rolling(2, min_periods = 1).apply(_update_min)
df
```
